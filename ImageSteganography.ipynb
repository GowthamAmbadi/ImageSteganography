```python
import os
import shutil
from PIL import Image
import random
import matplotlib.pyplot as plt
from steganography import message_to_binary, hide_data, show_data
```

```python
# Optional: folder for cover images (use any local path)
dataset_path = os.path.join(os.getcwd(), "images")
os.makedirs(dataset_path, exist_ok=True)
```

```python
# Core encode/decode logic is in steganography.py (imported above)
```

```python
def preview_image(image_path):
    """Display image before encoding."""
    img = Image.open(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title("Selected Cover Image")
    plt.show()

def encode_image():
    image_name = input("Enter image path (e.g., ./image.png or full path): ")
    secret_msg = input("Enter the secret message: ")
    output_path = input("Enter the image name to save (e.g., stego_image.png): ")
    password = input("Password (optional, press Enter for none): ").strip() or None
    preview_image(image_name)
    hide_data(image_name, secret_msg, output_path=output_path, password=password)
    print("‚úÖ Message encoded into:", output_path)

def decode_image():
    image_name = input("Enter the stego image path to decode (e.g., stego_image.png): ")
    password = input("Password (optional, if you set one when encoding): ").strip() or None
    try:
        print("Decoded message:", show_data(image_name, password=password))
    except ValueError as e:
        print("Error:", e)
```

```python
def main():
    while True:
        print("\n--- IMAGE STEGANOGRAPHY ---")
        print("1. Encode a message")
        print("2. Decode a message")
        print("3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            encode_image()
        elif choice == '2':
            decode_image()
        elif choice == '3':
            print("üëã Exiting...")
            break
        else:
            print("‚ö†Ô∏è Invalid option, please try again.")

if __name__ == "__main__":
    main()
```
